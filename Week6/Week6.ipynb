```python
#initialize all the packages and modules needed
import pandas as pd
import statsmodels.tools.tools as stattools
from sklearn.tree import DecisionTreeClassifier, export_graphviz
```

```python
#Save the income data to variable y for simplicity
adult_tr = pd.read_csv('adult_ch6_training.csv')
y=adult_tr[['Income']]
```

```python
y
```

Utilize the Cart Model implemented found within the SKlearn package
========
make use of a dummy variable to for the marital status using
categorical() command

```python
#creates an array that contains the Marital Status
import numpy as np
mar_np = np.array(adult_tr['Marital status'])
mar_cat = pd.get_dummies(mar_np)
```

```python
mar_cat
```

```python
#Save multiple variable from the Categorical() method
mar_cat_pd = pd.DataFrame(mar_cat)
x = pd.concat((adult_tr[['Cap_Gains_Losses']], mar_cat_pd), axis =1)
```

```python
mar_cat_pd
```

```python
x
```

Create Labels for the graph to help visualize the CART model

```python
x_names = ["Cap_Gains_Losses", "Divorced", "Married", "Never-married",
           "Separated", "Widowed"]

y_names = ["<50k", ">50K"]
```

```python
#create the CART Algorithm
#the DecisionTreeClassifier() sets up parameters for decision tree
#criterion="gini" specifies the use of CART model
#max_leaf_nodes indicats the CART tree to at most the specified number
cart01 = DecisionTreeClassifier(criterion="gini", max_leaf_nodes=5).fit(x,y)
cart01
```

```python
#use export_graphvic() to output the cart01
data = export_graphviz(cart01, out_file=None, feature_names=x_names, class_names=y_names)
```

```python
#use precit() commad to obtain classifications of t he income variable
predIncomeCART = cart01.predict(x)
predIncomeCART
```
